```python
# Importing the pandas package
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read in data
aqi_17 = pd.read_csv('https://aqs.epa.gov/aqsweb/airdata/daily_aqi_by_county_2017.zip', compression= 'zip')
aqi_18 = pd.read_csv('https://aqs.epa.gov/aqsweb/airdata/daily_aqi_by_county_2018.zip', compression= 'zip')

# Combining the aqi_17 and aqi_18 data frames (or stacking them) using pd.concat()
aqi = pd.concat([aqi_17, aqi_18])

# Simplify column names
aqi.columns= (aqi.columns.str.lower().str.replace(' ', '_'))

# Filtering to only the Santa Barbara County data in the aqi_sb data set and assigning it to aqi_sb
aqi_sb = aqi.loc[aqi['county_name'] == "Santa Barbara"]

#Dropping the columns 'state_name', 'county_name', 'state_code', and 'county_code' from the aqi_sb data frame
aqi_sb = aqi_sb.drop(['state_name', 'county_name', 'state_code', 'county_code'], axis=1)

# Updating the date column of the aqi_sb data frame to be in datetime64 using pd.to_datetime
aqi_sb.date= pd.to_datetime(aqi_sb['date'])

# Setting the date column to be the index of the aqi_sb data frame using .set_index
aqi_sb= aqi_sb.set_index('date')

# Calculate AQI rolling average over 5 days
rolling_average= aqi_sb['aqi'].rolling('5D').mean()

# Adding a new column called 'five_day_average' to the aqi_sb data frame using .assign() and the rolling_average variable
aqi_sb= aqi_sb.assign(five_day_average= rolling_average)

# Creating the plot of Air Quality Index Over Time in Santa Barbara County with both the aqi column and the five_day_averages column

plt.plot(aqi_sb.index, aqi_sb["aqi"], label="AQI") # Plotting a line graph of the AQI data by date 
plt.xlabel("Date")  # Adding X-axis label
plt.ylabel("Air Quality Index")  # Adding Y-axis label
plt.plot(aqi_sb.index, aqi_sb['five_day_average'], label="Five Day Average AQI") # Adding the five day average line 
plt.title("Air Quality Index Over Time in Santa Barbara County")  # Adding title
plt.xticks(rotation = 45) # Rotating the x-axis labels 
plt.legend(loc="upper left") # Adding in a legend for the lines
plt.show()
```

In the next cell we will calculate an average over a [rolling window](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.rolling.html) using the `rolling()`method for `pandas.Series`:

- `rolling()` is a lazy method, so we need to specify what we want to calculate over each window before it does something. 
- in this example we use the aggregator function `mean()` to calculate the average over each window
- the parameter '5D' indicates we want the window for our rolling average to be 5 days.

Store your answer in the `rolling_average` variable. You should have two columns in your series, `date` and the averages.
